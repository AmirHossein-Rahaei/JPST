```python
import pandas as pd
import numpy as np
np.random.seed(23)
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import RepeatedKFold
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Load the data
df = pd.read_csv("dataset//My-Data.csv")
X = df.drop('rich amine loading', axis=1)  # Replace 'target_column_name' with the name of your target column
y = df['rich amine loading']

# Initialize the model and cross-validated RepeatedKFold
model = RandomForestRegressor()
cv = RepeatedKFold(n_splits=10, n_repeats=10, random_state=42)

# Lists to store predicted and actual values
predicted_values = []
actual_values = []

# Loop through the folds and collect predictions and actual values
for train_index, test_index in cv.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    predicted_values.extend(y_pred)
    actual_values.extend(y_test)
    
r2 = r2_score(actual_values, predicted_values)
# Fit a linear regression model
lr_model = LinearRegression()
lr_model.fit(np.array(actual_values).reshape(-1, 1), predicted_values)

# Get the regression line points
line = np.linspace(min(actual_values), max(actual_values), 100)
regression_line = lr_model.predict(line.reshape(-1, 1))

# Plot the actual values vs. the predicted values with the regression line
plt.figure(figsize=(8, 6))
sns.regplot(x=actual_values, y=predicted_values, scatter_kws={'facecolors': 'none', 'edgecolors': 'darkblue', 'alpha': 0.5}, line_kws={'color': 'black'})
plt.xlabel("Industrial rich amine loading")
plt.ylabel("Predicted riich amine loading")
#plt.title("Comparison between simulation results and actual result of rich amine loading for amine absorption column",fontsize=10, weight="bold")
plt.grid(True)

plt.text(0.05, 0.95, f"R = {r2:.2f}", transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')



# Calculate and print R2 score for training with noise
from sklearn.metrics import r2_score
r2 = r2_score(actual_values, predicted_values)
print("R2 score for training with noise: ", 100 * r2)
plt.savefig("Figures//6.png", dpi=1000)
plt.show()
```

```python
import pandas as pd
import numpy as np
np.random.seed(23)
from sklearn.model_selection import RepeatedKFold
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Load the data
df = pd.read_csv("dataset//My-Data.csv")
X = df.values
X_t, y_t = X[:, :-1], X[:, -1]

# Initialize the model and cross-validated RepeatedKFold
cv = RepeatedKFold(n_splits=10, n_repeats=10, random_state=42)
model = RandomForestRegressor()

# Lists to store evaluation metrics
r2_scores = []
mse_scores = []
mae_scores = []
rmse_scores = []

# Loop through the cross-validation folds and collect metrics
for train_index, test_index in cv.split(X_t):
    X_train, X_test = X_t[train_index], X_t[test_index]
    y_train, y_test = y_t[train_index], y_t[test_index]

    # Train the model on the training data
    model.fit(X_train, y_train)

    # Make predictions on the test data
    y_pred = model.predict(X_test)

    # Calculate evaluation metrics on the test set
    r2_scores.append(r2_score(y_test, y_pred))
    mse_scores.append(mean_squared_error(y_test, y_pred))
    mae_scores.append(mean_absolute_error(y_test, y_pred))
    rmse_scores.append(np.sqrt(mean_squared_error(y_test, y_pred)))

# Calculate average metrics over all cross-validation folds
avg_r2 = np.mean(r2_scores)
avg_mse = np.mean(mse_scores)
avg_mae = np.mean(mae_scores)
avg_rmse = np.mean(rmse_scores)

# Print the average metrics on the test set
print("Average R-squared on test set:", avg_r2)
print("Average Mean Squared Error (MSE) on test set:", avg_mse)
print("Average Mean Absolute Error (MAE) on test set:", avg_mae)
print("Average Root Mean Squared Error (RMSE) on test set:", avg_rmse)
```
